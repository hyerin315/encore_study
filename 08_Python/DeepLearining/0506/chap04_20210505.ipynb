# 추측통계의 기본
* 통계적 추정

## 모집단과 표본

* 추측통계에서는 관측 대상 전체의 통계적 성질을 그 관측 대상의 일부분만 사용하여 추측합니다. 
* 추측하고 싶은 관측 대상 전체를 모집단(population) → 모집단 구하기 어려움(대부분 표본)
* 추측에 사용하는 관측 대상의 일부분을 표본(sample) 
* 모집단에서 표본을 골라내는 일을 표본추출(sampling)
* 골라낸 표본의 수를 표본 크기 또는 샘플 사이즈라고 합니다. 
* 표본을 바탕으로 계산한 평균이나 분산, 상관계수 등을 표본통계량이라고 합니다. 
* 모집단의 평균이나 분산, 상관계수 등을 모수라고 합니다. 
* 표본의 표본평균, 모집단의 평균을 모평균 등으로 부르는 경우도 있습니다.

* 표본을 대상으로 표본평균이라는 표본통계량을 사용해서 학교 전체의 평균 점수인 모평균을 추측
* 표본평균으로 모평균을 추측할 때, 표본평균은 모평균의 추정량(estimator)라고 합니다.
* 실제로 표본의 데이터를 이용하여 계산한 결과를 추정값(estimate)이라고 합니다. 
* 표본오차 : 모집단에서 복수의 표본을 무작위로 추출하면 곤란한 문제가 생긴다. 모집단의 성질이 표본별로 달라져 버리는데 이 성질을 표본오차라 한다. 이 표본오차를 극복하는 것이 확률론이다. 

### 표본의 추출방법
* 전교생 400명의 시험 점수가 모집단
* A 학생이 학교에서 우연히 만난 학생 20명의 시험 점수가 표본으로 추출된 점수들 
* 이 표본을 대상으로 표본 평균이라는 표본통계량을 사용해서 학교 전체의 평균 점수인 모평균이라는 모수를 추측하는 것

---------------
* 표본은 학교 안에서 우연히 만난 20명의 점수였음 
* 이 우연히 만난 것은 추측통계에서 매우 중요한 역할을 함. 
* 우연히 만난 20명에게 시험 결과를 듣는 것과 A 학생과 친하게 지내는 친구 20명에게 시험 결과를 듣는 것은 어떤 차이가 있을까? 
----------------------
* A 학생은 자신과 친하게 지내는 친구 20명에게 표본을 얻은 상황을 생각해 볼 수 있습니다.
* A 학생의 친구는 아마 A학생과 같은 3학년이 대부분일 것
* 3학년은 1학년이나 2학년 학생에 비해 높은 점수를 얻을 것으로 생각되므로, A 학생이 자기 친구들에게 물어서 얻은 표본평균은 전교생의 평균 점수보다 높아질 것이다. 이 방법은 전교생의 평균 점수가 올바르게 추측되었다고 할 수 없음 
* 이를 위해 A 학생은 표본을 '우연히 만나는' 학생들에게 모음으로써 표본추출이 편향(bias)되지 않도록 미리 작선을 세운다. 이처럼 임의로 표본을 추출하는 방법을 무작위추출(random sampling)이라고 함


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%precision 3
%matplotlib inline
```

```python
df = pd.read_csv('./data/ch4_scores400.csv')
scores = np.array(df['score'])
scores[:10]
```

```python
type(scores)
```

* 임의로 표본을 추출하는 방법을 무작위추출(random sampling)이라고 합니다.

```python
np.random.choice([1, 2, 3], 3)
```

* 우연히 만나는 방법을 썼을 때 같은 사람에게 여러 번 시험 결과를 들을 가능성도 있습니다. 
* 같은 사람에게 같은 질문을 여러 번 하면 의아한 눈으로 쳐다보겠지만, 추출 방법의 관점에서 보면 전혀 문제가 되지 않습니다. 
* 이처럼 여러 차례 동일한 표본을 선택하는 추출 방법을 복원추출(sampling with replacement)라고 합니다.
* 동일한 표본은 한 번만 선택하는 방법을 비복원추출(sampling without replcement)

```python
np.random.choice([1, 2, 3], 3, replace=False)
```

```python
np.random.choice([1, 2, 3], 3, replace=True)
```

```python
np.random.seed(0) #랜덤 고정
np.random.choice([1, 2, 3], 3)
```

```python
np.random.seed(0)
sample = np.random.choice(scores, 20)

sample.mean()
```

```python
scores.mean()
```

```python
for i in range(100):
    sample = np.random.choice(scores, 20)
    print(f'{i+1}번째 무작위 추출로 얻은 표본평균', sample.mean())
```

## 확률 모델
* 불확정성을 수반한 현상을 해석하기 위해 확률(probability)를 사용합니다.
* 확률을 사용한 무작위추출 혹은 주사위를 모델링 한 것을 확률 모형(probability model)이라고 합니다.

### 확률의 기본

### 확률분포

* 결과를 알아맞힐 수는 없지만, 취하는 값과 그 값이 나올 확률이 결정되어 있는 것을 확률변수(random variable)이라고 합니다.
* 주사위를 굴릴 때까지 어떤 눈이 나올지 알 수 없지만, 주사위를 굴리면 눈은 하나씩 결정 
* 이와 같이 확률변수의 결과를 관측하는 것을 시행(trial)이라고 하고, 시행에 의해 관측되는 값을 실현값(realization)이라고 합니다.
* 눈이 1이나 눈이 홀수 처럼 시행 결과로 나타날 수 있는 일을 사건(event)이라고 합니다.
* 눈이 1처럼 더 세부적으로 분해할 수 없는 사건을 근원사건(elementary event)이라고 합니다.

* 상호배반(mutually exclusive)이라는 것은 각 사건이 동시에 일어날 수 없다는 이야기

#### 확률분포
* 확률분포(probability distiribution)란 확률변수가 어떻게 움직이는지를 나타낸 것

* 불공정한 주사위를 확률분포를 구하기 위해서 아래처럼 설정

```python
dice = [1, 2, 3, 4, 5, 6]
prob = [1/21, 2/21, 3/21, 4/21, 5/21, 6/21] #확률
```

```python
np.random.choice(dice, p=prob)
```

```python
num_trial = 100
sample = np.random.choice(dice, num_trial, p=prob)
sample
```

```python
np.histogram(sample, bins=6, range=(1, 7))
```

```python
freq, _ = np.histogram(sample, bins=6, range=(1, 7))
dice_df = pd.DataFrame({'frequency':freq,
              'relative frequency':freq / num_trial},
              index = pd.Index(np.arange(1, 7), name='dice'))
```

```python
dice_df['relative frequency'].sum()
```

```python
prob
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
# 실제의 확률분포를 가로선으로 표시
ax.hlines(prob, np.arange(1, 7), np.arange(2, 8), colors='gray')
# 막대 그래프의 [1.5, 2.5, ..., 6.5]에 눈금을 표시
ax.set_xticks(np.linspace(1.5, 6.5, 6))
# 주사위 눈의 값은 [1, 2, 3, 4, 5, 6]
ax.set_xticklabels(np.arange(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')
plt.show()
```

```python
num_trial = 100000
sample = np.random.choice(dice, size=num_trial, p=prob)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
ax.hlines(prob, np.arange(1, 7), np.arange(2, 8), colors='gray')
ax.set_xticks(np.linspace(1.5, 6.5, 6))
ax.set_xticklabels(np.arange(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')
plt.show()
```

* 표본 크기가 커지면, 실제의 분포에 수렴

```python
sample = np.random.choice(scores, 10000)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=100, range=(0, 100), density=True)
ax.set_xlim(20, 100)
ax.set_ylim(0, 0.042)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()
```

* 표본 하나하나가 확률변수이므로, 표본들의 평균으로 계산되는 표본평균도 확률변수
* 여기서 무작위추출로 표본 크기가 20인 표본을 추출하여 표본평균을 계산하는 작업을 1000번 수행해서, 그 결과를 히스토그램으로 그려 표본평균이 어떻게 분포되는지 살펴보겠습니다.

```python
sample_means = [np.random.choice(scores, 40).mean()
                for _ in range(100000)]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample_means, bins=100, range=(0, 100), density=True)
# 모평균을 세로선으로 표시
ax.vlines(np.mean(scores), 0, 1, 'gray')
ax.set_xlim(50, 90)
ax.set_ylim(0, 0.13)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show() #정규분포 그래프
```
