## 1. 하나의 데이터프레임 만들기

```python
df1 = pd.read_csv("./data/Subway/Subway_201601.csv", encoding='cp949')
subway = df1

for x in range(2,12) :    
    if x > 9:
        df2 = pd.read_csv("./data/Subway/Subway_2016{}.csv".format(x), encoding='cp949')
        df2 = df2.drop([df2.index[0]])
        subway = pd.concat([subway, df2], ignore_index=True)
    else:
        df2 = pd.read_csv("./data/Subway/Subway_20160{}.csv".format(x), encoding='cp949')
        df2 = df2.drop([df2.index[0]])
        subway = pd.concat([subway, df2], ignore_index=True)
```
