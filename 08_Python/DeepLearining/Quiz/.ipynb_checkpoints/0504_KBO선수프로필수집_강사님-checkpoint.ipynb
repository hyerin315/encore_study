```python
def playerID(id_):
    play_dict = {}
    url = "https://www.koreabaseball.com/Record/Player/HitterDetail/Basic.aspx?playerId={}".format(id_)
    r = requests.get(url)
    #print (r.url)
    for x in BS(r.text).find("div",\
                class_ = "player_basic").findAll("li"):
        rt = [y.strip() for y in x.text.strip().split(":")] #콜론으로 분리 후 공백 없애기
        play_dict[rt[0]] = rt[1] #콜론의 왼쪽 key값, 오른쪽은 value 값
    play_dict['구단명'] = BS(r.text).find("h4", id="h4Team").text
    return play_dict
```

```python
kbo = kbo[['구단명', '선수명', '등번호', '생년월일', '포지션', '신장/체중', '경력', '입단 계약금', '연봉', '지명순위',
       '입단년도']]
```

```python
kbo['연봉'] = kbo['연봉'].apply(lambda x : int(x.replace("달러", "")) * 1125  if x.find('달러') > -1 else x)
kbo['연봉'] = kbo['연봉'].apply(lambda x : int(x.replace("만원", "")) * 10000 if str(x).find("만원") > -1 else x)
```

```python
kbo_salary = kbo[kbo['연봉'] != ''].copy()
```

```python
kbo_salary['연봉'] = kbo_salary['연봉'].astype(np.int64)
kbo_salary.sort_values(by=['연봉'], ascending=False)
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
kbo_salary.groupby('구단명')['연봉'].mean().sort_values(ascending=False)
```
