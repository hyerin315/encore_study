### 일별시세크롤링

```python
import requests
import pandas as pd
from selenium import webdriver
import time
```

```python
url = "https://finance.naver.com/item/sise_day.nhn?code=005930&page={}"
```

```python
#나의 접속정보를 가져옴
header = {'user-agent':
          'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36'
}
```

```python
#나의 접속정보를 변장해 정보를 받아옴(그냥 들어가면 정보를 받아올 수 없게 해놨음)
r = requests.get(url, headers = header)
r.text
```

```python
#10번 반복하며 정보 가져오기
total = []
for x in range(1, 11):
    r = requests.get(url.format(x), headers = header)
    total.append(pd.read_html(r.text)[0].dropna())
```

```python
pd.concat(total) #데이터 프레임 결합
```

```python
stock = pd.concat(total)
stock.reset_index(drop = True, inplace = True) #인덱스 리셋
```

```python
pd.read_html(r.text)[0].dropna() #테이블 형태를 그대로 가져오기
```

```python
#내가 했던 것(오류발생)
page_cnt = 10
total_df = []
for x in range(1, page_cnt + 1):
    print("{}페이지의 정보 수집을 시작합니다 ===================".format(x))
    
    print('page','/',x)
    
    total_df.append(pd.read_html(r.text)[0])
    print(total_df)
    
    x += 1
    if(x % 10 == 1):
        driver.find_element_by_link_text("다음").click()
    else:
        url = 'http://finance.naver.com/item/sise_day.nhn?code=005930&page={}'.format(x)
        r = requests.get(url.format(x), headers = header)

    time.sleep(2)
```

```python
#데이터 하나하나 가져와 담기
#방법1 - get 방식
url = "https://finance.naver.com/item/sise_day.nhn?code={}&page=1"

kospi_total = []
for idx, code in enumerate(krx_df[krx_df['mkt_tp_nm'.upper()] == 'kospi'.upper()]['ISU_SRT_CD'].values):
    if idx % 100 == 0 and idx > 0 :break
    #payload['symbol'] = code
    r = requests.get(url.format(code), data=payload, headers=header)
    tmp_df = pd.read_html(r.text)[0].dropna()
    tmp_df['code'] = code
    kospi_total.append(tmp_df)
```

```python
#방법2 - post 방식
kospi_total = []
for idx, code in enumerate(krx_df[krx_df['mkt_tp_nm'.upper()] == 'kospi'.upper()]['ISU_SRT_CD'].values): #kospi가 포함된 데이터에서 단축코드 갖고오기
        if idx % 100 == 0 and idx > 0:break
        payload['symbol'] = code
        r = requests.post(krx_post_url, data=krx_payload, headers=header)
        break
    
        tmp_df = pd.read_html(r.text)[0].dropna()
        tmp_df['code'] = code
        kospi_total.append(tmp_df)
```

```python
pattern = re.compile(p)
```

```python
rt = [x.replace('"', "").strip() for x in pattern.findall(r.text.split("\n")[6])[0].split(",")]
```
