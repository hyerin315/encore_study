```python
names = []
births = []
genders = []
money = []

#데이터 읽어와 리스트에 담기
for roots, dirs, files in os.walk("./data/이름_생년_성별_10000"):
    for file in files:
        name, birth, gender = file.split("_")
        names.append(name)
        births.append(birth)
        genders.append(gender[0])
        
        #파일 열어서 안의 내용 읽어오기
        with open(roots + "/" + file, "r") as f:
            money.append(f.read())
```

```python
import pandas as pd
```

```python
#데이터프레임 만들기
df = pd.DataFrame({"이름" : names, "생년월일" : births, "성별" : genders, "금액" : money})
df
```

```python
from datetime import date, datetime, timedelta
```

```python
#데이트타임패키지 사용
datetime.strptime("20210410", "%Y%m%d") #사용법 확인
```

```python
df['생년월일'] = df['생년월일'].apply(lambda x : datetime.strptime(x, "%Y%m%d"))
df
```

```python
#weekday() 사용법
a = datetime.strptime("20210510", "%Y%m%d")
a.weekday() #0은 월요일
```

```python
#월요일에 태어난 사람만 추출
df[df['생년월일'].apply(lambda x : x.weekday() == 0)]
```

```python
#주말에 태어난 사람 제외하여 추출
df[df['생년월일'].apply(lambda x : x.weekday() != 5,6)]
```

```python
#100일잔치 날짜 뽑기
df['생년월일'].apply(lambda x : x + timedelta(days = 100))
```

```python
df['성별'] = df['성별'].apply(lambda x : '남' if x in ['1', '3'] else '여')
df
```

```python
df['금액'] = df['금액'].apply(lambda x : int(x.replace(",", "")))
df
```

```python
df['생년월일'].apply(lambda x : ages(x))
```

```python
def ages(x):
    age = 2021 - x.year + 1
    
    if age <=20:
        return '10대'
    elif age > 20 and age <= 30:
        return '20대'
    elif age > 30 and age <= 40:
        return '30대'
    elif age > 40 and age <= 50:
        return '40대'
    else:
        return '50대'
```

```python
df['연령대'].apply(lambda x : ages(x))
```
