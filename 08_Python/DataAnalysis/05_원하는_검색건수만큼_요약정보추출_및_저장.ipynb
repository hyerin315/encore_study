```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
import math

query_txt = input("크롤링할 키워드를 입력하세요 : ")
cnt = int(input("크롤링할 건수를 입력하세요 : "))
page_cnt = math.ceil(cnt/10) #크롤링할 전체 페이지 수

#검색결과를 저장할 경로
f_name = input("txt파일명과 경로를 지정하세요(예: C:\\LHR\\python\\notebook\\data\\test4.txt) :")
fc_name = input("csv파일명과 경로를 지정하세요(예: C:\\LHR\\python\\notebook\\data\\csv4.csv) :")
fx_name = input("xlsx파일명과 경로를 지정하세요(예: C:\\LHR\\python\\notebook\\data\\xlsx4.xlsx) :")

path = "C:/LHR/python/datadown/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://korean.visitkorea.or.kr")
time.sleep(2)

try:
    driver.find_element_xpath('//*[@id="chkForm01"]').click()
except:
    print("코로나 창이 없습니다")

element = driver.find_element_by_id('inp_search')
element.send_keys(query_txt)
element.send_keys('\n')

time.sleep(2)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

print('\n')
print("="*70)
print("요청하신 검색 건수는 {}건이고".format(cnt))
print("총 {}페이지의 게시글 수집을 진행합니다".format(page_cnt))
print("="*70)
```

```python
no = 1
no2 = []
contents2 = []
tags2 = []
```

```python
s_time = time.time()
for x in range(1, page_cnt + 1): #1페이지부터 반복 시작 (끝까지 돌기 위해 +1 꼭 추가)
    print("%s 페이지 내용을 수집합니다==================" %x)
    print("\n")
    
    time.sleep(2)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    contents_list = soup.find('ul', 'list_thumType type1')
    
    for i in contents_list:
        no2.append(no)
        print("번호 : ", no)
        
        try:
            contents = i.find('div', 'tit').get_text() #있으면 콘텐츠 값 가져오기
        except AttributeError:
            contents = "내용 없음"
            contents2.append(contents)
            print("내용 : ", contents.strip()) #빈공간이 있다면 없애기
        else:
            contents2.append(contents)
            print("내용 : ", contents.strip())
            
        try :
            tag = i.find('p','tag_type').get_text()
        except AttributeError :
            tag = '태그정보가 없음'
            tags2.append(tag)
            print("태그 : ", tag.strip())
        else:
            tags2.append(tag)
            print("태그 : ", tag.strip())
            
        if no == cnt:
            break
            
        no += 1
        
    time.sleep(2)
    
    x += 1 #여기서의 x는 for문의 x -> 다음페이지를 클릭하기 위해 +1
    if x == page_cnt + 1:
        break
        
    if x == 6:
        driver.find_element_by_link_text('다음').click()
    else:
        driver.find_element_by_link_text("{}".format(x)).click()
```

```python
korea = pd.DataFrame()
korea['번호'] = no2
korea['내용'] = contents2
korea['태그'] = tags2
```

```python
f = open(f_name, 'a', encoding='UTF-8')
f.write(str(no2))
f.write(str(contents2))
f.write(str(tags2))
f.close()
```

```python
korea.to_csv(fc_name, encoding='UTF-8-sig', index=False) #index는 데이터 프레임이 갖고 있기 때문에 따로 만들지 않음
korea.to_excel(fx_name , index=False)
```

```python
e_time = time.time()
t_time = e_time - s_time
```

```python
print('\n')
print("=" * 80)
print("크롤링을 요청한 총 {}건수 중에서 {}건의 데이터를 수집완료했습니다".format(cnt, no))
print("총 소요시간 {}초 입니다".format(round(t_time, 1))) #너무 길게 출력돼서 소숫점 첫째자리까지 출력하도록 round 사용
print("파일저장 완료 - txt 파일명: {}".format(f_name))
print("파일저장 완료 - csv 파일명: {}".format(fc_name))
print("파일저장 완료 - xlsx 파일명: {}".format(fx_name))
print("=" * 80)

driver.close()
```
