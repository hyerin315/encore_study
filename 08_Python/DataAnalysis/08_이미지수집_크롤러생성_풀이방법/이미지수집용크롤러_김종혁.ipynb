```python
#환경설정
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
import math
import os
import urllib
import urllib.request

f_dir = input('사진을 저장할 폴더를 지정하세요(예: C:\\study\\python\\notebook\\data\\) : ')
query_txt = '사진저장'

#파일을 저장할 폴더 생성
now = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' %(now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)
s_short = '%04d-%02d-%02d' %(now.tm_year, now.tm_mon, now.tm_mday)

#입력받은 경로+현재 날짜_사진저장이라는 폴더 생성
os.makedirs(f_dir+s+'-'+query_txt)
f_result_dir = f_dir+s+'-'+query_txt
#폴더명 출력
print(f_result_dir)

#크롬 드라이버를 사용하여 웹 브라우저 실행
path = "C:/study/python/datadown/chromedriver.exe"
driver = webdriver.Chrome(path)
s_time = time.time()
# driver.get('https://korean.visitkorea.or.kr/detail/rem_detail.do?cotid=d8db847f-b262-4048-90a4-6aa4b20bee6e&con_type=10100')
driver.get('https://korean.visitkorea.or.kr/detail/cs_detail_cos.do?cotid=1923378b-3e45-4ae9-8610-c8235636ba5e&big_category=A01&mid_category=A0100&big_area=39')
time.sleep(2)  # 페이지가 모두 열릴 때 까지 2초 기다립니다.
driver.maximize_window()

## Step 4.  스크롤다운 함수를 생성한 후 실행합니다.스크롤다운 함수를 생성한 후 실행합니다.
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(1)
    
# scroll_down(driver) 

#본문의 사진 정보를 가져와 지정된 폴더에 저장하기
file_no = 0
ftitle=[]
img_src2=[]

try:
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')

    img_src=soup.find('div','box_txtPhoto').find_all('img')
    for i in img_src:
        img_src1=i['src'] #이미지 태그의 src 속성값(파일경로및파일명) 가져오기
        img_src2.append(img_src1)
except:
    print('이미지 게시글이 아닙니다.')

try:
    for x in range(1,5):
        driver.find_element_by_xpath("""//*[@id="contents"]/div[2]/div[6]/div[1]/div[1]/div/ul/li[{}]/a""".format(x)).click()
        time.sleep(2)
        html=driver.page_source
        soup=BeautifulSoup(html,'html.parser')

        title = soup.find('ul','swiper-wrapper cosList').find_all('div','tit')[x-1].get_text()
        tag = soup.find('div',"cos_cont active")
        tag2 = tag.find('div','info_area pc').find_all('a')
        for i in tag2:
            ftitle.append(title)
            style = i['style'].split('(')[1].split(')')[0]
            img_src2.append(style)
except:
    print('코스 게시글이 아닙니다.')

os.chdir(f_dir+s+'-'+query_txt) #다운되는 곳의 경로를 생성된 특정 폴더의 경로로 이동

for i in range(0,len(img_src2)):
    print("%s 번째 이미지 저장중입니다========" %file_no)
    try:
        urllib.request.urlretrieve(img_src2[i], str(s_short)+' '+str(ftitle[i])+' '+str(file_no)+'.jpg')
    except:
        continue
    file_no += 1
    if(file_no%3==0):
        file_no = 0
    time.sleep(0.5)

#요약정보 출력
e_time=time.time()
t_time=e_time-s_time

print("=" *70)
print("총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("총 저장 건수는 %s 건 입니다 " %file_no)
print("파일 저장 경로: %s 입니다" %f_result_dir)
print("=" *70)
 
driver.close( )
```
