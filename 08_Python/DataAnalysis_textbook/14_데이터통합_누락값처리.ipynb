```python
sample_1 = pd.read_excel("data2/sample_1.xlsx",
                         header = 1,
                         skipfooter = 2, 
                         usecols = 'A:C')
sample_1
```

```python
code_master = pd.read_excel('data2/sample_codemaster.xlsx')
code_master
```

```python
sample_1_code = pd.merge(left = sample_1,
                         right = code_master,
                         how = 'left',
                         left_on = '국적코드',
                         right_on = '국적코드')
sample_1_code
```

```python
sample_1_code = pd.merge(left = sample_1,
                         right = code_master,
                         how = 'inner',
                         left_on = '국적코드',
                         right_on = '국적코드')
sample_1_code
```

```python
sample_1_code['기준년월'] = '2019-11'
sample_1_code
```

```python
sample_2 = pd.read_excel('data2/sample_2.xlsx',
                         header = 1,
                         skipfooter = 2,
                         usecols = 'A:C')
sample_2['기준년월'] = '2019-12'
```

```python
sample_2_code = pd.merge(left = sample_2,
                          right = code_master, 
                          how = 'left',
                          left_on = '국적코드',
                          right_on = '국적코드')
sample_2_code
```

```python
sample = sample_1_code.append(sample_2_code, ignore_index = True)
sample
```

```python
sample.to_excel('data2/sample.xlsx', index = False) #인덱스 번호 제외하고 저장
```
