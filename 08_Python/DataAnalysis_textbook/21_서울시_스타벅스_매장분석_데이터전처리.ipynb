```python
import pandas as pd
```

```python
#스타벅스 파일 읽어오기
seoul_sb = pd.read_excel('data4/seoul_sb_list.xlsx', header = 0) #header = 0부터 읽어오겠다
seoul_sb.head()
```

```python
#스타벅스 주소에서 시군구명 추출
sgg_name = []
for address in seoul_sb['주소']:
    sgg = address.split()[1] #sgg : 배열, split() : 빈 값으로 나누겠다
    sgg_name.append(sgg)
    
sgg_name
```

```python
#시군구명 열 생성 및 삽입
seoul_sb['시군구명'] = sgg_name
seoul_sb.info()
```

```python
#엑셀로 저장
seoul_sb.to_excel('data4/seoul_starbucks_list.xlsx', index = False)
```

```python
#시군구 목록 읽어오기
seoul_sgg = pd.read_excel('data4/seoul_sgg_list.xlsx')
seoul_sgg.head()
```

```python
#서울시 스타벅스 매장 목록 데이터 불러오기
seoul_starbucks = pd.read_excel('data4/seoul_starbucks_list.xlsx')
seoul_starbucks.head()
```

```python
#시군구별 스타벅스 매장 수 세기
#index는 시군구명으로 매장명을 보여줄 건데, count라는 통계량으로 보여주겠다
starbucks_sgg_count = seoul_starbucks.pivot_table(
                                                  index = '시군구명',
                                                  values = '매장명',
                                                  aggfunc = 'count',
                                                  ).rename(columns = {'매장명':'스타벅스_매장수'})
starbucks_sgg_count.head()
```

```python
#서울시 시군구 목록 데이터에 스타벅스 매장 수 데이터 병합
#왼쪽을 기준으로 병합하고, 둘다 갖고 있는 시군구명을 기준으로 병합하겠다
seoul_sgg = pd.merge(seoul_sgg, starbucks_sgg_count, how = 'left', on = '시군구명')
seoul_sgg.head()
```

```python
#서울시 시군구별 인구통계 데이터 불러오기
seoul_sgg_pop = pd.read_excel('data4/sgg_pop.xlsx')
seoul_sgg_pop.head()
```

```python
#서울시 시군구 목록 데이터에 서울시 시군구별 인구통계 데이터를 병합
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_pop, how = 'left', on = '시군구명')
seoul_sgg.head()
```

```python
#서울시 시군구 목록 데이터에 서울시 시군구별 사업체 수 통계 데이터를 병합
seoul_sgg_biz = pd.read_excel('data4/sgg_biz.xlsx')
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_biz, how = 'left', on = '시군구명')
seoul_sgg.head()
```

```python
#병합 결과를 엑셀파일로 저장
seoul_sgg.to_excel('data4/seoul_sb_final.xlsx', index = False)
```
