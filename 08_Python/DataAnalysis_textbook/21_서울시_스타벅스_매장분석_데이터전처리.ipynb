```python
import pandas as pd
```

```python
#스타벅스 파일 읽어오기
seoul_sb = pd.read_excel('data4/seoul_sb_list.xlsx', header = 0) #header = 0부터 읽어오겠다
seoul_sb.head()
```

```python
#스타벅스 주소에서 시군구명 추출
sgg_name = []
for address in seoul_sb['주소']:
    sgg = address.split()[1] #sgg : 배열, split() : 빈 값으로 나누겠다
    sgg_name.append(sgg)
    
sgg_name
```

```python
#시군구명 열 생성 및 삽입
seoul_sb['시군구명'] = sgg_name
seoul_sb.info()
```

```python
#엑셀로 저장
seoul_sb.to_excel('data4/seoul_starbucks_list.xlsx', index = False)
```
