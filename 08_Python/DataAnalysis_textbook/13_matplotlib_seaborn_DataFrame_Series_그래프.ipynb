```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
tips = tips.rename({'sex':'gender'}, axis='columns')
tips.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1) #1행, 1열에 1번째 그래프를 그린다
                               #숫자는 자동으로 만들어짐
```

```python
#여러개의 그래프를 비교하고 싶을 때 subplot 사용
axes1.hist(tips['total_bill'], bins=10) #total_bill의 값을 나눠서 보여줄 bins : 막대 값(구간)
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Frequency")
fig
```

```python
plt.hist(tips['total_bill'], bins = 5)
plt.title("Histogram of Totla bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
```

```python
plt.scatter(tips['total_bill'], tips["tip"])
plt.title("Scatterplot of total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.boxplot(#불린 추출
    [tips[tips['gender'] == 'Female']['tip'], #tips에서 True인 값만 추출
     tips[tips['gender'] == 'Male']['tip']],
    labels = ['Female', 'Male']
)
plt.title("Boxplot of Tips by Gender")
plt.xlabel("Gender")
plt.ylabel("Tip")
plt.show()
```

```python
def recode_gender(gender):
    if gender == 'Female':
        return 0
    else:
        return 1
```

```python
tips['gender_color'] = tips['gender'].apply(recode_gender) #tips 개수만큼 recode_gender의 매개변수로 들어가면서 브로드캐스팅이 이루어진다
tips.head()
```

```python
plt.scatter(
    x = tips['total_bill'], #지불금액
    y = tips['tip'], #팁,
    s = tips['size'] * 10, #테이블당 인원수 * 10 (크게 보여주기 위해)
    c = tips['gender_color'],
    alpha = 0.5 #1이 기본값 (투명도)
)

plt.title('Total Bill vs Tip Colored by gender and Sized by Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
