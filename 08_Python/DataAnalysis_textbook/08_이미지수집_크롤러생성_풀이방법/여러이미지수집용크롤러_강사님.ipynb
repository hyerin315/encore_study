```python
from bs4 import BeautifulSoup
from selenium import webdriver
import urllib.request
import urllib
import pandas as pd
import time
import math
import os
import random
```

```python
print("="*80)
print("예제:대한민국 구석구석 사이트의 여행지 정보 이미지 수집하기")
print("="*80)
query_txt = input("크롤링할 키워드는 무엇입니까?:")
cnt = int(input("크롤링할 건수는 몇건입니까?"))
page_cnt = math.ceil(cnt/10)
print("출력될 페이지수:",page_cnt)
```

```python
f_dir = input('결과파일을 저장할 폴더명을 쓰세요(C:\\study\\python\\notebook\\data\\):')
n = time.localtime()
s = '{}-{}-{}-{}-{}-{}'.format(n.tm_year,n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)
result_dir = f_dir+s+'-'+query_txt
os.makedirs(f_dir+s+'-'+query_txt)
result_dir
```

```python
s_time = time.time()
path = "c:/study/python/datadown/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get('https://korean.visitkorea.or.kr')
time.sleep(random.randrange(1,3))

try :
    driver.find_element_by_xpath("""//*[@id="chkForm01"]""").click()
except :
    print("코로나 창이 없습니다")
    
element  = driver.find_element_by_id('inp_search')
element.send_keys(query_txt)
element.send_keys("\n")

print("="*80)
print('크롤링할 키워드:',query_txt)
print('크롤링할 총 페이지 번호:',page_cnt)
print("="*80)
```

```python
no = 1
for x in range(1, page_cnt + 1):
    print("{}페이지 이미지 수집 시작합니다.=====================".format(x))
    for i in range(1,13):
        if no > cnt:
            break
     
   
        #제목이름의 폴더 생성
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        try :
            title = soup.select("""#listBody > ul > li:nth-child({}) > div.area_txt > div.tit > a"""\
                            .format(i))[0].get_text()

        except Exception as e:
            print(e)


        try:
            driver.find_element_by_xpath("""//*[@id="listBody"]/ul/li[{}]/div[2]/div[1]/a""".format(i)).click()
        except:
            driver.back()
            continue


        #상세페이지로 이동
        time.sleep(random.randrange(1,3)) 

        driver.maximize_window()

        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        img_src2 = []

        #이미지 주소추출

        try:
            img_src = soup.find('div','box_txtPhoto').find_all('img')
            for i in img_src:
                img_src1 = i['src']
                img_src2.append(img_src1)
            
        except Exception as e:
            try:
                for y in range(1,5):
                    driver.find_element_by_xpath("""//*[@id="contents"]/div[2]/div[6]/div[1]/div[1]/div/ul/li[{}]/a""".format(y)).click()

                    time.sleep(2)

                    html = driver.page_source
                    soup = BeautifulSoup(html,'html.parser')
                    tag = soup.find('div','cos_cont active')
                    tag2 = tag.find('div','info_area pc').find_all('a')

                    for i in tag2:
                        style = i['style'].split('(')[1].split(')')[0] #style속성안에 괄호안에있는거
                        img_src2.append(style)

            except Exception as e:
                print(e);print('유형3')


        #이미지 다운로드 폴더 생성
        if len(img_src2) != 0:
            print("title",title)
            os.makedirs(result_dir+'\\'+title)
            os.chdir(result_dir+'\\'+title)

        #이미지 다운로드 실행
        file_no = 0
        print("이미지 갯수:",len(img_src2))
        for i in range(0,len(img_src2)):
            try:
                urllib.request.urlretrieve(img_src2[i],str(file_no)+'.jpg')
            except:
                continue
            file_no += 1
            time.sleep(0.5)
            print("{} 번째 이미지가 저장중 입니다.=====================".format(file_no))


        driver.back()
        time.sleep(2)
        no +=1

    if x > page_cnt + 1:
        break
        
    x += 1
              
    time.sleep(1)     
    if(x % 5 == 1):
        driver.find_element_by_link_text("다음").click()
    else:
        driver.find_element_by_link_text("{}".format(x)).click()
        
    time.sleep(2)
```

```python
e_time = time.time()
t_time = e_time - s_time


print("="*80)
print('총 소요시간은 {}초 입니다.'.format(round(t_time,1)))
print('파일 저장 경로 : {} 입니다.'.format(result_dir))
print("="*80)


driver.close()
```
