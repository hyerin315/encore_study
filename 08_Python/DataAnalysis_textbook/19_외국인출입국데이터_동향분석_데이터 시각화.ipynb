```python
import matplotlib.pyplot as plt
```

```python
condition = (df['국적'] == '중국')
df_filter = df[condition] #df에서 True 인 것만 가져옴
df_filter.head()
```

```python
#사이즈 - 가로 : 12, 세로 : 4
plt.figure(figsize = (12, 4))

#df_filter : 중국에 대한 데이터 / x축 : '기준년월' / y축 : '관광'
plt.plot(df_filter['기준년월'], df_filter['관광'])

#그래프의 타이틀, x축, y축 이름 달기
plt.title('중국 국적의 관광객 추이')
plt.xlabel('기준년월')
plt.ylabel('관광객수')

#x축 눈금값 설정
plt.xticks(['2010-01', '2010-01', '2011-01', '2012-01', '2013-01', '2014-01', '2015-01', '2016-01', '2017-01', '2018-01', '2019-01'])

#그래프 표현
plt.show()
```

```python
#우리나라 방문하는 외국인 관관객 top5 리스트 만들기 
cntry_list = ['중국', '일본', '대만', '미국', '홍콩']
```

```python
for cntry in cntry_list:
    #국적 관광객만 추출하기 
    condition = (df['국적'] == cntry) #국적 데이터로 변환
    df_filter = df[condition]
    plt.figure(figsize = (12, 4))
    
    #df_filter : 국가별 데이터 / x축 : '기준년월' / y축 : '관광'
    plt.plot(df_filter['기준년월'], df_filter['관광'])

    #그래프의 타이틀, x축, y축 이름 달기
    plt.title(f'{cntry}의 관광객 추이')
    plt.xlabel('기준년월')
    plt.ylabel('관광객수')

    #x축 눈금값 설정
    plt.xticks(['2010-01', '2010-01', '2011-01', '2012-01', '2013-01', '2014-01', '2015-01', '2016-01', '2017-01', '2018-01', '2019-01'])
```

```python
#히트맵 그래프 그리기
import seaborn as sns 

#str.slice() 함수를 이용해 년도, 월 컬럼 만들기
df['년도'] = df['기준년월'].str.slice(0, 4) #0위치에서 4바로 전까지
df['월'] = df['기준년월'].str.slice(5, 7) #5위치에서 7ㅂ바로 전까지
df.head()
```

```python
#중국 관광객 필터링하기
condition = (df['국적'] == '중국')
df_filter = df[condition]
df_filter.head()
```

```python
#pivot_table() 함수 활용하기
df_pivot = df_filter.pivot_table(values = '관광',
                                 index = '년도',
                                 columns = '월')
df_pivot
```

```python
#그래프 크기 설정
plt.figure(figsize = (16, 10))

#히트맵 그래프 그리기
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')

#그래프 타이틀 달기
plt.title('중국 관광객 히트맵')

#그래프 보이기
plt.show()
```

```python
#반복문으로 top5 국가 히트맵 그래프 그리기
for cntry in cntry_list:
    condition = (df['국적'] == cntry)
    df_filter = df[condition]
    
    df_pivot = df_filter.pivot_table(values = '관광',
                                     index = '년도',
                                     columns = '월')
    
    plt.figure(figsize = (16, 10))
    
    sns.heatmap(df_pivot, annot = True, fmt = '.1f', cmap = 'rocket_r')
    
    plt.title('{} 관광객 히트맵'.format(cntry))
    
    plt.show()
```
